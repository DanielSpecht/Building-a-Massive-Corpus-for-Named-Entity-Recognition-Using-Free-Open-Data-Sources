# Total de referências

```python
# build dictionary 'num references' to 'num article'
references_stats = {}
for article in client['wikipedia_ner_corpus']['wikipedia_article'].find():
    article_refrences = article['reference_count']
    
    if not article_refrences in references_stats:
        references_stats[article_refrences] = 0
    references_stats[article_refrences] += 1
```

```python
# dict to array
total_size = sum(references_stats.values())
ref_array = np.zeros((total_size))

i = 0
for count_refs, count_arts in references_stats.items():
    
    for j in range(count_arts):
        ref_array[i] = count_refs
        i+=1
```

```python
#sort
sorted_keys = sorted(references_stats.keys())
```

```python
# checkpoints list will be filled with [start, end, total] where total is the percentage of the values in the range
# init -> end, the treshold defines a value to efine a range
checkpoints = []
treshold = 10
previous = 0
current = 0
curr_init = None

for ref_count in sorted_keys:
    
    if not curr_init:
        curr_init = ref_count
    
    art_count = references_stats[ref_count]
    percent = (float(art_count)/float(total_size))*100
    #print(ref_count,'---', art_count,'---',"%02f"%(percent))
    
    current += percent
    
    if current - previous >= 10 or ref_count == sorted_keys[len(sorted_keys) - 1]:
        checkpoints.append([curr_init, ref_count, current - previous])
        previous = current
        curr_init = None

# sanity test
print(checkpoints)
print(sum([c[2] for c in checkpoints] ))
```

Build
sentences
    - total words -> how many sents have that many words
    - total annotated -> how many sents
    - total predicted -> how many sents

tokens
predicted/annotated/total
    - by type:
    - size -> how many
    - total

```python
# Sent Structure
# 
# 'pred': {total:int,  }
#
def add2info(key, val, dictionary):
    dictionary = {}

dataset_infos = {'total':0,
                 'sizes':{},
                 
                 'ents_by_sent':{'annotated': {},
                                       'predicted': {}},
                 
                 'predicted':{'total':0,
                              'PER':{'sizes':{},'total':0},
                              'LOC':{'sizes':{},'total':0},
                              'ORG':{'sizes':{},'total':0}},
    
                 'annotated':{'total':0,
                              'PER':{'sizes':{},'total':0},
                              'LOC':{'sizes':{},'total':0},
                              'ORG':{'sizes':{},'total':0}}}

i = 0
for art in client['wikipedia_ner_corpus']['wikipedia_article'].find():
    for sent in art['sentences']:
        
        sent_infos = sent['infos']
        
        # add to sentences count
        dataset_infos['total'] += 1
        
        # add to sizes count, the quantity of sentences with that amount of tokens
        tokens_count = len(sent['tokens'])
        if not tokens_count in dataset_infos['sizes']:
            dataset_infos['sizes'][tokens_count] = 0
        dataset_infos['sizes'][tokens_count] += 1

        
        for key in ['predicted', 'annotated']:
            
            # entities by sent, the quantity of sentences with that amount of entities of that origin
            if not sent_infos[key]['count'] in dataset_infos['ents_by_sent'][key]:
                dataset_infos['ents_by_sent'][key][sent_infos[key]['count']] = 0
            dataset_infos['ents_by_sent'][key][sent_infos[key]['count']] += 1
            
            # no entities here
            if sent_infos[key]['count'] == 0:
                continue
                
            # add to source count
            dataset_infos[key]['total'] += sent_infos[key]['count']
             
            # entity types count
            for kind in list(sent_infos[key]['type_count'].keys()):
                 dataset_infos[key][kind]['total'] += sent_infos[key]['type_count'][kind]
            
            # entity type length
            # e.g. {'PER':[1,3,5]}
            for kind in list(sent_infos[key]['lengths_by_type'].keys()):
                 
                 for size in sent_infos[key]['lengths_by_type'][kind]:
                     if not size in dataset_infos[key][kind]['sizes']:
                         dataset_infos[key][kind]['sizes'][size] = 0
                     dataset_infos[key][kind]['sizes'][size] += 1
    
    
    #i += 1
    #if i > 100:
    #    break    
```
